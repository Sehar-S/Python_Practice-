```python
# Numeric data type example: Integer
x = 4
type(x) # Check the data type of x
```

```python
x = 7.4
type(x) # Check the data type of x (float)
```

```python
y = 5 + 6j
type(y) # Check the data type of y (complex number)
```

## Change Data types

```python
var = 7 # Initialize an integer variable
y = float(var) # Convert integer to float
y # Display the float value
```

```python
z = int(y) # Convert float back to integer
z # Display the integer value
```

```python
c = complex(4, 7) # Create a complex number with real part 4 and imaginary part 7
c # Display the complex number
```

```python
# Boolean data type example: Comparison (False)
a = 6
b = 9
c = a > b # Check if 'a' is greater than 'b'
c # Display the boolean result
```

```python
# Boolean data type example: Comparison (True)
a = 6
b = 9
c = a < b # Check if 'a' is less than 'b'
c # Display the boolean result
```

```python
int(True) # Convert boolean True to integer (results in 1)
```

```python
int(False) # Convert boolean False to integer (results in 0)
```

```python
# String data type example
p = 'python' # Assign a string value to p
type(p) # Check the data type of p
```

```python
# Sequence type example: List
lst = [1, 2, 3, 4, 5] # Create a list of integers
type(lst) # Check the data type of lst
```

```python
# Sequence type example: Tuple
tup = (43, 56, 77, 88) # Create a tuple of integers
type(tup) # Check the data type of tup
```

```python
# Sequence type example: Range
r = range(18) # Create a range object from 0 to 17
type(r) # Check the data type of r
```

```python
r # Display the range object
```

```python
Display = list(r) # Convert the range object to a list
Display # Display the list
```

```python
# Mapping type example: Dictionary
d = {'bilal': 3456784, 'akram': 3456789, 'Ali': 3456780} # Create a dictionary with names and numbers
d # Display the dictionary
```

```python
d['akram'] # Access the value associated with the key 'akram'
```

```python
d.values() # Get all values from the dictionary
```

```python
d.keys() # Get all keys from the dictionary
```

```python
# Set data type example
s = {45, 66, 77, 'ali'} # Create a set with mixed data types
type(s) # Check the data type of s
```
