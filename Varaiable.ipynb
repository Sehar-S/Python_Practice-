### 5.2 Conditional Logic for Weather

```python
age =23
salary=60000
if age > 22 and salary > 50000:
    loan_status = "Approved"
else:
    loan_status = "Rejected"
print("Loan Status:", loan_status)
```

### 5.3 Multiple Conditions with Logical Operators (AND)

### 5.4 User Input with Conditional Logic

```python
marks=80
if marks >= 80:
    result ='A Grade'
elif marks >= 50: # This condition checks if marks are between 50 and 79
    result= 'B Grade'
else: # If marks are less than 50
    result = 'fail'
print('result is:',result)
```
